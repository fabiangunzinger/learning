```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import iris
```

```python
### Load and inspect data 

# background: http://www.dicook.org/files/jsm19/slides#1
iris = sns.load_dataset('iris')
iris.head()

iris.describe()

iris.species.unique()
```

```python
### Basic plots

sns.set()
sns.swarmplot(x='species', y='petal_length', data=iris)
plt.show()
```

```python
### ECDFs for sepal length 

def draw_ecdf(data, label_name):
    """Draw empirical CDF of data series. """
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1)/n
    pct = np.array([2.5, 25, 50, 75, 97.5])
    pct_data = np.percentile(data, pct)
    
    plt.plot(x, y, marker='.', linestyle='none', label=label_name.title())
    plt.plot(pct_data, pct/100, marker='D', linestyle='none')
    plt.ylabel('ECDF')

plt.figure(figsize=(10, 7))
plt.title('Iris sepal lengths')

plt.subplot(2, 2, 1)
for species in iris.species.unique():
    data = iris.sepal_length[iris.species == str(species)]
    draw_ecdf(data, species)
plt.xlabel('Sepal length (cm)')
plt.legend()
    
plt.subplot(2, 2, 2)
for species in iris.species.unique():
    data = iris.sepal_width[iris.species == str(species)]
    draw_ecdf(data, species)
plt.xlabel('Sepal width (cm)')
    
plt.subplot(2, 2, 3)
for species in iris.species.unique():
    data = iris.petal_length[iris.species == str(species)]
    draw_ecdf(data, species)
plt.xlabel('Petal length (cm)')

plt.subplot(2, 2, 4)
for species in iris.species.unique():
    data = iris.petal_width[iris.species == str(species)]
    draw_ecdf(data, species)
plt.xlabel('Petal width (cm)')



for l in 'sepal', 'petal':
    print(l)
```
